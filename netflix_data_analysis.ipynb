Netflix Data Analysis using Python and Pandas




```python
import pandas as pd
import matplotlib.pyplot as plt
```

1. Load the Dataset

```python
df = pd.read_csv("/content/netflix_titles.csv")
```

3. Data Cleaning

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.fillna("unknown",inplace=True)
```

```python
df.isnull().sum()
```

4. Exploratory Data Analysis (EDA)

```python
df["type"].value_counts()
```

4.1 Movies vs TV Shows

```python
df["type"].value_counts().plot(kind="bar")
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.savefig("movies_vs_tv_shows.png")
plt.show()
```

4.2 Top 10 Countries

```python
df["country"].value_counts().head(10)
```

```python
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries with Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.savefig("top_10_countries.png")
plt.show()
```

4.3 Top Genres

```python
df["listed_in"].value_counts().head(10)
```

```python
df["listed_in"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.savefig("top_genres.png")
plt.show()
```

4.4 Content Ratings

```python
df["rating"].value_counts()
```

```python
df["rating"].value_counts().plot(kind="bar")
plt.title("Netflix Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.savefig("ratings.png")
plt.show()
```

4.5 Release Year Analysis

```python
df["release_year"].value_counts().head(10)
```

```python
df["release_year"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Release Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.savefig("release_years.png")
plt.show()
```

Conclusion
